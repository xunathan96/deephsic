```python
# Import ratinabox 
import ratinabox
from ratinabox.Environment import Environment
from ratinabox.Agent import Agent
from ratinabox.Neurons import *
import pandas as pd 
```

```python
# Run a very simple simulation 
env = Environment()
agent = Agent(env)
PCs = PlaceCells(agent)
for i in range(int(60 / agent.dt)):
    agent.update()
    PCs.update()
```

```python
# Export data into a dataframe 
pd.DataFrame(agent.history)
```
